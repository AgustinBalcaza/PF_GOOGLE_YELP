```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Especifica la ruta del archivo Parquet
parquet_file_path = 'sample.parquet'

# Lee el archivo Parquet en un DataFrame de Pandas
df = pd.read_parquet(parquet_file_path)
```

```python
# Eliminando columnas irrelevantes
df1 = df.drop(['Time', 'Address', 'Price', 'Rating', 'Num_of_reviews', 'User_id'], axis=1)
```

```python
# ObtÃ©n el Ã­ndice de la columna
column_index = df1.columns.get_loc('Business name')

# Mueve la columna a la izquierda (posiciÃ³n 0)
df1.insert(0, 'Business name', df1.pop('Business name'))
```

```python
# borrando duplicados
data = df1.drop_duplicates()
```

```python
# Reseteando el indice 
data.reset_index(drop=True, inplace=True)
```

```python
# cambiando tipos de datos
data_1['Latitude'] = data_1['Latitude'].astype(str)
data_1['Longitude'] = data_1['Longitude'].astype(str)
data_1['Avg_rating'] = data_1['Avg_rating'].astype(str)
```

```python
# Combinar los valores de las columnas en una sola columna
data_1['features'] = data_1[['Latitude', 'Longitude', 'Category', 'Avg_rating', 'State']].apply(lambda x: ', '.join(x), axis=1)

data_1.drop(['Latitude', 'Longitude', 'Category', 'Avg_rating', 'State', 'Review'],axis=1,inplace=True)
```

```python
def recomendacion_hoteles(Business_name):
    # cargar los datasets que se van a utilizar para dos dataframes distintos
    data1 = pd.read_csv('data_hoteles.csv')
    
    # Obtener el feature del hospedaje conocido
    features_a_comparar = data1[data1['Business name'] == Business_name]['features'].values[0]

    # Filtrar el DataFrame para hospedajes con el mismo feature (excluyendo el hospedaje conocido)
    df_filtrado = data1[data1['Business name'] != Business_name]
    
    # Inicializar el vectorizador TF-IDF y aplicarlo a los features de los hospedajes
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(df_filtrado['features'])

    # Vectorizar el feature conocido
    vector_feature_conocido = tfidf_vectorizer.transform([features_a_comparar])

    # Calcular la similitud del coseno entre el feature conocido y los feature de los demÃ¡s hospedajes
    similarity_scores = cosine_similarity(vector_feature_conocido, tfidf_matrix)

    # Obtener los Ã­ndices de los 5 hospedajes mÃ¡s similares
    indices_hospedajes_similares = similarity_scores.argsort()[0][-5:][::-1]

    # Obtener los nombres de los hospedajes recomendados utilizando los Ã­ndices
    hospedajes_recomendados = df_filtrado.loc[indices_hospedajes_similares]['Business name'][1:5].tolist()

    # mensaje que indica el negocio del hospedaje original y los negocio de hospedajes recomendados
    mensaje = f"Si te gustÃ³ el hospedaje {Business_name}, tambiÃ©n te pueden gustar:"

    result_dict = {
        'mensaje': mensaje,
        'hospedajes recomendados': hospedajes_recomendados
    }
    
    return result_dict
```
