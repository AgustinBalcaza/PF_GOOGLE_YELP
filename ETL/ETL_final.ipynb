# ETL Final de Google Maps y Yelp

Aquí realizaremos el ETL final donde concatenaremos los datos de los 5 estados seleccionados de los datasets de Google Maps y le haremos las transformaciones finales a los datos proporcionados por Yelp

```python
import pandas as pd
```

Para empezar leemos los datasets de cada estado y los guardamos dentro de una variable

```python
ca_hotels = pd.read_parquet('ca_hotels.parquet')
```

```python
ca_hotels
```

```python
ca_hotels['state'] = 'California'
```

```python
ca_hotels
```

```python
fl_hotels = pd.read_parquet('fl_hotels.parquet')
```

```python
fl_hotels['state'] = 'Florida'
```

```python
nv_hotels = pd.read_parquet('nv_hotels.parquet')
```

```python
nv_hotels['state'] = 'Nevada'
```

```python
ny_hotels = pd.read_parquet('ny_hotels.parquet')
```

```python
ny_hotels['state'] = 'Nueva York'
```

```python
tx_hotels = pd.read_parquet('tx_hotels.parquet')
```

```python
tx_hotels['state'] = 'Texas'
```

```python
tx_hotels
```

Ahora concatenamos todos los dataframes en uno solo

```python
df = pd.concat([ca_hotels, fl_hotels, nv_hotels, ny_hotels, tx_hotels], axis=0, ignore_index=True)
```

```python
df
```

```python
google_maps = df
```

Eliminamos columnas que no son de nuestro interés

```python
google_maps.drop(['text', 'address', 'description', 'price'], axis=1, inplace=True)
```

```python
google_maps
```

Realizaremos una transformación de la columna 'time' para extraer solamente el mes y el año:

```python
google_maps['time'] = google_maps['time'].astype(str)
```

```python
google_maps
```

```python
google_maps['year'] = google_maps['time'].apply(lambda x: x[:4])
google_maps['month'] = google_maps['time'].apply(lambda x: x[5:7])
```

```python
google_maps
```

Finalmente eliminamos la columna 'time' y convertimos las nuevas columnas 'month' y 'year' a tipo entero

```python
google_maps.drop('time', axis=1, inplace=True)
```

```python
google_maps
```

```python
google_maps['year'] = google_maps['year'].astype(int)
google_maps['month'] = google_maps['month'].astype(int)
```

Guardamos el dataset

```python
google_maps.to_parquet('Google_Maps.parquet', index=False)
```

Ahora importamos el dataset de Yelp (luego de haberle realizado transformaciones) y lo guardamos en una variable

```python
review = pd.read_parquet('review.parquet')
```

```python
review
```

Eliminamos columnas que no son de nuestro interés

```python
review.drop(['business_id', 'city', 'state', 'latitude', 'longitude', 'user_id'], axis=1, inplace=True)
```

```python
review
```

Separamos las categorías

```python
reviews_yelp = review.assign(categories=review['categories'].str.split(', ')).explode('categories').reset_index(drop=True)
```

```python
reviews_yelp
```

Realizamos el mismo procedimiento anterior para transformar la columna 'date' y extraer solamente el mes y el año

```python
reviews_yelp['year'] = reviews_yelp['date'].apply(lambda x: x[:4])
reviews_yelp['month'] = reviews_yelp['date'].apply(lambda x: x[5:7])
```

```python
reviews_yelp
```

```python
reviews_yelp.drop('date', axis=1, inplace=True)
```

```python
reviews_yelp['year'] = reviews_yelp['year'].astype(int)
reviews_yelp['month'] = reviews_yelp['month'].astype(int)
```

```python
reviews_yelp
```

Guardamos en formato Parquet

```python
reviews_yelp.to_parquet('Reviews_Yelp.parquet', index=False)
```
