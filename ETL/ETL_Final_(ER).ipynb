Generamos un código para calcular el análisis de sentimiento basado en el resultado y el rating dividido entre 5 para asignarle el valor de 2 (positivo), 1 (neutral) o 0 (negativo), en el caso de las reseñas que no contienen texto (solamente dejaron rating de estrellas), se consideró como positivo los ratings de 4 y 5, como neutral los ratings de 3 estrellas y como malo los ratings de 1 y 2 estrellas

```python
def calcular_sentimiento(row):
    if pd.notnull(row['analisis']):
        resultado = row['analisis'] + row['rating'] / 5
        if resultado > 1.5:
            return 2
        elif 1 <= resultado <= 1.5:
            return 1
        else:
            return 0
    elif pd.notnull(row['rating']):
        if row['rating'] >= 4:
            return 2
        elif row['rating'] == 3:
            return 1
        else:
            return 0
    else:
        return None
```

```python
google_maps['sentiment analysis'] = google_maps.apply(calcular_sentimiento, axis=1)
```
